## load weights

```python
pretrain_weights = []

for k, v in yolo_pretrain.items():
    if len(v.shape) > 1:
        pretrain_weights.append(np.ascontiguousarray(np.transpose(v, (3, 2, 0, 1))))
    else:
        pretrain_weights.append(v)
    print(k, '=>', v.shape)
```

```python
from tqdm import tqdm
mark_index = 0

for m in tqdm(darknet53.modules()):
    if isinstance(m, torch.nn.Conv2d):
        assert list(m.weight.size()) == list(pretrain_weights[mark_index].shape)
        m.weight.data = torch.from_numpy(pretrain_weights[mark_index])
        if m.bias is not None:
            mark_index += 1
            assert list(m.bias.shape) == list(pretrain_weights[mark_index].shape)
            m.bias.data = torch.from_numpy(pretrain_weights[mark_index])
        mark_index += 1
    if isinstance(m, torch.nn.BatchNorm2d):
        assert list(m.weight.shape) == list(pretrain_weights[mark_index].shape)
        m.weight.data = torch.from_numpy(pretrain_weights[mark_index])
        mark_index += 1
        
        assert list(m.bias.shape) == list(pretrain_weights[mark_index].shape)
        m.bias.data = torch.from_numpy(pretrain_weights[mark_index])
        mark_index += 1
        
        assert list(m.running_mean.shape) == list(pretrain_weights[mark_index].shape)
        m.running_mean.data = torch.from_numpy(pretrain_weights[mark_index])
        mark_index += 1
        
        assert list(m.running_var.shape) == list(pretrain_weights[mark_index].shape)
        m.running_var.data = torch.from_numpy(pretrain_weights[mark_index])
        mark_index += 1
```

```python
model_state = darknet53.state_dict()

state_dict = {'model': model_state,
                      'optim': None,
                      'epoch': 0,
                      'steps': 0,
                      'loss': float('inf')}

torch.save(state_dict, '/home/dk/Desktop/darknet53.pkl')
```
